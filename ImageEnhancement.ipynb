```python
import glob, os
from os import walk, getcwd
import os.path
from PIL import Image
import shutil


out = "/mydrive/Yolov4/darknet/data/Enhanced_exdark/"
path = "/mydrive/Yolov4/darknet/data/Allanno/"

os.chdir(path)

wd = getcwd()

os.chdir(wd)

txt = []

for file in glob.glob("*"):

    if os.path.isfile(out):
        print(file)
        print("File exists")
        txt.append(file)
    else:   
        print(file)
        shutil.copy(file, out)
        txt.append(file)



print(len(txt))
print("Done :)")
```
